```python
from snownlp import SnowNLP
import pandas as pd
```

```python
def show_result(coment):
    s = SnowNLP(coment)
    return s.sentiments
```

```python
path = "data/低端.csv"
df = pd.read_csv(path)
print(df.head())
```

```python
columns_delete = ["IMEI"]
df = df.drop(columns = columns_delete)
df["类别"] = df["类别"].str.replace("_x0000_","",regex = True)
df["描述"] = df["描述"].str.replace("_x0000_","",regex = True)
print(df)
print(df["类别"].unique())
```

```python
df.replace("",pd.NA,inplace = True)
df = df.dropna()
print(df)
```

```python
df["sentiment"] = df["描述"].apply(show_result)
print(df)
```

```python
df.to_csv('data/低端_sentiments.csv', index=False)
```
